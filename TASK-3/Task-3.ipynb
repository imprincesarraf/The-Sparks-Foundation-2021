## BIVARIATE ANALYSIS

```python
df.corr()
```

```python
import seaborn as sns
sns.pairplot(df.select_dtypes(exclude=["object"]))
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataplot = sn.heatmap(df.corr(),annot=True)
plt.show()
```

#### Clearly the discount and profit are highly negatively corelated
###### Values closer to 0 implies that there is no linear dependency between the variables
###### Values closer to 1 implies that all data points lie on a line for which Y increases as X increases
###### Values closer to -1 implies that all data points lie on a line for which Y increases as X decreases


```python
plt.figure(figsize=(1,1))
df.groupby('Category')['Profit','Sales','Discount'].sum().plot(kind="bar")
plt.show()
```

##### From the above graph clearly sales of  furniture is  good but still profit is lowest

```python
plt.figure(figsize=(1,1))
df.groupby('Sub-Category')['Sales','Discount','Profit'].sum().plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(5,5))
df.groupby('Sub-Category')['Sales'].sum().plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(5,5))
df.groupby('Sub-Category')['Profit'].sum().plot(kind='bar')
plt.show()
```

##### From the above graph we can see that:
###### Sales of tables is high but it is producing loss
###### There is negligible loss from Bookcases and Supplies
###### There is negligible sales and profit from fasteners
###### Copiers are giving highest profit


```python
plt.figure(figsize=(1,1))
df.groupby('Region')['Sales','Discount','Profit'].sum().plot(kind='bar')
plt.show()
```

```python
sns.barplot(x=df.Region,y=df.Sales)
```

```python
sns.barplot(x=df.Region,y=df.Profit)
```

```python
sns.barplot(x=df.Region,y=df.Profit,hue=df.Category)
```

#### Furniture is making loss from Central region

#### From the above graph we can see that:

```python
df1=df.groupby('State')['Profit','Sales'].sum()
df1.plot.bar(figsize=(20,10))
```

```python
df1=df.groupby('State')['Profit'].sum()
df1.plot.bar(figsize=(20,10))
```

#### From the above graph we can see that:
##### 1.Sales  and Profit in California and New York is more
##### 2. States like Arizona,Colorado,Flordia,IIinois,North Carolina,Ohio,Pennsylvania,Tennessee,Texas are making loss


```python
df2=df.groupby('State')['Discount'].sum()
plt.xlabel("States")
plt.ylabel("Discount")
df2.plot.bar(figsize=(20,10))
```

#### Texas, Arizona,Colorado,Flordia,IIinois,North Carolina,Ohio,Pennsylvania,Tennessee these states are giving good discounts which results loss
